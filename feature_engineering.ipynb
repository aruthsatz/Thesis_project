This 'dataset_filter' function grabs previous 40 games for both teams
and combines them into a dataset. 

```python
def dataset_filter(home_code, away_code, game_date):
    dataset = test.loc[(test['date'] < game_date)]

    team1_home = dataset[dataset["home"] == home_code].tail(20)
    team1_away = dataset[dataset["away"] == home_code].tail(20)
    team1_df = pd.concat([team1_home, team1_away])

    team2_home = dataset[dataset["home"] == away_code].tail(20)
    team2_away = dataset[dataset["away"] == away_code].tail(20)
    team2_df = pd.concat([team2_home, team2_away])

    combined_data = pd.concat([team1_df, team2_df])
    # both the seasons and test datasets have the same index
    # below I am saving the index column to a list and applying it to the seasons dataset so that it returns the same data
    
    combined_data_index = list(combined_data.index)
    combined_seasons_data = seasons.iloc[combined_data_index]
    # 11/15 i changed seasons to seasons_std to return a standardized dataset
    
    return combined_seasons_data
```

```python
def date_field_preprocessing(file_date):
    ## This method deletes days from the 'date' field
    file_date_list = list(file_date)

    new_date_list = []
    for t in file_date:
        date_num = str(t).replace(' days', '')
        new_date_list.append(date_num)
        
        
    new_date_series = pd.Series(new_date_list)
    return new_date_series
```
